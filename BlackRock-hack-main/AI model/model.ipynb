```
!pip install yfinance

import yfinance as yf
import pandas as pd

top_200_tickers = [
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'FB', 'BRK-B', 'JNJ', 'JPM', 'V',
    'NVDA', 'DIS', 'PYPL', 'PG', 'UNH', 'MA', 'HD', 'VZ', 'ADBE', 'NFLX',
    'INTC', 'CMCSA', 'KO', 'PEP', 'MRK', 'ABT', 'PFE', 'CSCO', 'XOM', 'NKE',
    'T', 'CVX', 'ORCL', 'DHR', 'LLY', 'MCD', 'WMT', 'BAC', 'CRM', 'QCOM',
    'WFC', 'AVGO', 'TXN', 'COST', 'NEE', 'MDT', 'HON', 'TMO', 'BMY', 'UNP',
    'AMGN', 'IBM', 'LOW', 'MMM', 'GE', 'ACN', 'UPS', 'MS', 'AMAT', 'BA',
    'SCHW', 'SBUX', 'LIN', 'LMT', 'CAT', 'BLK', 'RTX', 'SPGI', 'CVS', 'TGT',
    'INTU', 'ISRG', 'PLD', 'SYK', 'BKNG', 'NOW', 'ZTS', 'CHTR', 'DE', 'MDLZ',
    'CB', 'MU', 'ANTM', 'CL', 'MO', 'CCI', 'USB', 'ADP', 'FIS', 'EL', 'GILD',
    'CI', 'SO', 'DUK', 'GM', 'APD', 'TFC', 'ICE', 'ITW', 'PNC', 'C', 'BDX',
    'NSC', 'NSC', 'NOC', 'MCO', 'REGN', 'MMC', 'AON', 'EW', 'VRTX', 'HUM',
    'TMUS', 'AEP', 'TROW', 'ADI', 'CSX', 'LRCX', 'ETN', 'D', 'WM', 'GD',
    'ECL', 'FISV', 'PSA', 'HCA', 'HES', 'KMB', 'ILMN', 'TEL', 'BSX', 'FDX',
    'BK', 'COF', 'PGR', 'ROP', 'LHX', 'IDXX', 'F', 'BAX', 'AIG', 'TRV',
    'SLB', 'PRU', 'APH', 'STZ', 'PPG', 'MPC', 'MAR', 'ROP', 'CTSH', 'HPQ',
    'A', 'EMR', 'KMI', 'SYY', 'SPG', 'TT', 'JCI', 'RMD', 'ROK', 'PSX', 'DFS',
    'MKC', 'PH', 'YUM', 'DLTR', 'ALL', 'EXC', 'PCAR', 'AVB', 'WBA', 'SRE',
    'FTV', 'SYF', 'BKR', 'CTAS', 'BXP', 'ALB', 'WEC', 'IFF', 'DGX', 'EA',
    'DHI', 'AME', 'FTNT', 'CNC', 'DTE', 'PAYX', 'KEYS', 'AWK', 'KR', 'STE',
    'SWK', 'WLTW', 'CHD', 'DOV', 'SBAC', 'ZBH', 'CE', 'RJF', 'VLO', 'LEN'
]


def get_company_statistics(ticker_symbol):

    ticker_data = yf.Ticker(ticker_symbol)

    stock_info = ticker_data.info
    sustainability_data = ticker_data.sustainability

    company_name = stock_info.get('shortName', 'N/A')
    peg_ratio = stock_info.get('pegRatio', 'N/A')
    enterprise_value_ebitda = stock_info.get('enterpriseToEbitda', 'N/A')
    enterprise_value_revenue = stock_info.get('enterpriseToRevenue', 'N/A')
    forward_pe = stock_info.get('forwardPE', 'N/A')
    trailing_pe = stock_info.get('trailingPE', 'N/A')
    esg_score = sustainability_data.loc['totalEsg'].values[0] if 'totalEsg' in sustainability_data.index else 'N/A'

    data = {
        'Company Name': company_name,
        'PEG Ratio (5yr expected)': peg_ratio,
        'Enterprise Value/EBITDA': enterprise_value_ebitda,
        'Enterprise Value/Revenue': enterprise_value_revenue,
        'Forward P/E': forward_pe,
        'Trailing P/E': trailing_pe,
        'Total ESG Risk Score': esg_score
    }

    return data

all_data = []

for ticker in top_200_tickers:
    try:
        company_data = get_company_statistics(ticker)
        all_data.append(company_data)
        print(f"Successfully fetched data for {ticker}")
    except Exception as e:
        print(f"Failed to fetch data for {ticker}: {e}")

statistics_df = pd.DataFrame(all_data)

print(statistics_df)

statistics_df.to_csv('top_200_companies_statistics.csv', index=False)
```

```
def suggest_better_company(user_companies_data, all_companies_data):
    suggestions = []

    for user_company_data in user_companies_data:
        user_company_name = user_company_data.get('Company Name', 'Unknown Company')

        for company in all_companies_data:
            rationale = []

            # PEG Ratio (5yr expected)
            if company['PEG Ratio (5yr expected)'] != 'N/A' and user_company_data['PEG Ratio (5yr expected)'] != 'N/A':
                if (company['PEG Ratio (5yr expected)'] < 1 and user_company_data['PEG Ratio (5yr expected)'] > 1):
                    rationale.append("PEG Ratio (5yr expected) is better")

            # Enterprise Value/EBITDA
            if company['Enterprise Value/EBITDA'] != 'N/A' and user_company_data['Enterprise Value/EBITDA'] != 'N/A':
                if (company['Enterprise Value/EBITDA'] < 10 and user_company_data['Enterprise Value/EBITDA'] > 10):
                    rationale.append("Enterprise Value/EBITDA is better")

            # Enterprise Value/Revenue
            if company['Enterprise Value/Revenue'] != 'N/A' and user_company_data['Enterprise Value/Revenue'] != 'N/A':
                if (company['Enterprise Value/Revenue'] < 3 and user_company_data['Enterprise Value/Revenue'] > 3):
                    rationale.append("Enterprise Value/Revenue is better")

            # Forward P/E
            if company['Forward P/E'] != 'N/A' and user_company_data['Forward P/E'] != 'N/A':
                if (company['Forward P/E'] < 20 and user_company_data['Forward P/E'] > 20):
                    rationale.append("Forward P/E is better")

            # Trailing P/E
            if company['Trailing P/E'] != 'N/A' and user_company_data['Trailing P/E'] != 'N/A':
                if (company['Trailing P/E'] < 20 and user_company_data['Trailing P/E'] > 20):
                    rationale.append("Trailing P/E is better")

            # Total ESG Risk Score
            if company['Total ESG Risk Score'] != 'N/A' and user_company_data['Total ESG Risk Score'] != 'N/A':
                if (company['Total ESG Risk Score'] < user_company_data['Total ESG Risk Score']):
                    rationale.append("Total ESG Risk Score is better")

            # If any rationale is found, add the suggestion
            if rationale:
                suggestions.append({
                    'User Company Name': user_company_name,
                    'Suggested Company': company['Company Name'],
                    'Rationale': ', '.join(rationale)
                })

    return suggestions

user_companies = [
    {
        'Company Name': 'User Company 1',
        'PEG Ratio (5yr expected)': 1.5,
        'Enterprise Value/EBITDA': 12,
        'Enterprise Value/Revenue': 4,
        'Forward P/E': 25,
        'Trailing P/E': 22,
        'Total ESG Risk Score': 20
    },
    {
        'Company Name': 'User Company 2',
        'PEG Ratio (5yr expected)': 2.0,
        'Enterprise Value/EBITDA': 15,
        'Enterprise Value/Revenue': 5,
        'Forward P/E': 30,
        'Trailing P/E': 25,
        'Total ESG Risk Score': 60
    }
]

all_companies = [
    {
        'Company Name': 'Company A',
        'PEG Ratio (5yr expected)': 0.8,
        'Enterprise Value/EBITDA': 8,
        'Enterprise Value/Revenue': 2.5,
        'Forward P/E': 18,
        'Trailing P/E': 17,
        'Total ESG Risk Score': 40
    },
    {
        'Company Name': 'Company B',
        'PEG Ratio (5yr expected)': 0.5,
        'Enterprise Value/EBITDA': 9,
        'Enterprise Value/Revenue': 2.0,
        'Forward P/E': 19,
        'Trailing P/E': 18,
        'Total ESG Risk Score': 35
    },

]

def plot_stock_chart(ticker_symbol, start_date='2020-01-01', end_date='2024-01-01'):

    ticker_data = yf.Ticker(ticker_symbol)
    hist = ticker_data.history(start=start_date, end=end_date)

    plt.figure(figsize=(10, 5))
    plt.plot(hist.index, hist['Close'], label='Close Price')
    plt.title(f'{ticker_symbol} Stock Price')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.grid(True)
    plt.show()


#CHANGE 'ALL_COMPANIES'
suggestions = suggest_better_company(user_companies, all_companies)
for suggestion in suggestions:
    print(f"User Company: {suggestion['User Company Name']}")
    print(f"Suggested Company: {suggestion['Suggested Company']}")
    print(f"Rationale: {suggestion['Rationale']}\n")
    plot_stock_chart('Suggested Company')
    import matplotlib.pyplot as plt


```
